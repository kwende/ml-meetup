# The goal

the goal is to implement the scaled dot product attention idea and test it out using randomly generated vectors for the input sequence. 

### Define the class

```python
import numpy as np
import scipy as sp

batch_size = 64
sequence_length = 10

key_size = 64
query_size = 64
value_size = 64

keys = np.random.random((sequence_length, batch_size, key_size))
queries = np.random.random((sequence_length, batch_size, query_size))
values = np.random.random((sequence_length, batch_size, value_size))
```

```python
class ScaledDotProductAttention():
    def __init__(self, **kwargs):
        pass

    def __call__(self, keys, queries, values):
        dk = keys.shape[2]
        scaled_dot_products = np.matmul(queries, keys) / dk
        weight_matrix = sp.special.softmax(scaled_dot_products)
        return np.matmul(weight_matrix, values)
```

```python
scaledDP = ScaledDotProductAttention()

scaledDP(keys, queries, values)
```
